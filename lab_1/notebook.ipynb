```python
from ultralytics import YOLO
import os
```

```python
import torch
print(torch.cuda.is_available())
print(torch.version.cuda)
```

```python
model = YOLO('yolov8s.pt')

model.info()
print(model.names)
```

```python
images_path = os.path.abspath('./coco_subset/images/train/')
yaml_path = os.path.abspath('./coco_subset/coco.yaml')
```

```python
results = model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    name='yolov8_custom'
)
```

```python
metrics = model.val()
print(metrics)
```

```python
print(model.names)
```

```python
best_model = YOLO('runs/detect/yolov8_custom/weights/best.pt')

predict_results = best_model.predict(source='coco_subset/images/train', save=True, conf=0.5)

for i, r in enumerate(predict_results):
    r.save(filename=f"/test_predict/{i}.png")
```
